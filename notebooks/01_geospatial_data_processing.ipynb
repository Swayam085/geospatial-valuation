```python
import pandas as pd

# Load Dataset
df = pd.read_csv("../data/raw/kc_house_orig.csv")

print("="*60)
print("Dataset Shape")
print(df.shape)

print("="*60)
print("Columns")
print(df.columns.tolist())

print("="*60)
print("Dataset Info")
df.info()

print("="*60)
print("Missing Values")
print(df.isnull().sum())

print("="*60)
print("Duplicate Rows")
print(df.duplicated().sum())

print("="*60)
print("Statistical Summary")
print(df.describe())

print("="*60)
print("First 5 Rows")
display(df.head())

print("="*60)
print("Last 5 Rows")
display(df.tail())
```

```python
import matplotlib.pyplot as plt

print("="*60)
print("Duplicate Rows:", df.duplicated().sum())

print("="*60)
print("Missing Values")
print(df.isnull().sum())

print("="*60)
print("Latitude Range")
print(df["lat"].min(), "to", df["lat"].max())

print("="*60)
print("Longitude Range")
print(df["long"].min(), "to", df["long"].max())

print("="*60)
print("Price Range")
print(df["price"].min(), "to", df["price"].max())

# Price Distribution
plt.figure(figsize=(8,5))
plt.hist(df["price"], bins=50)
plt.title("House Price Distribution")
plt.xlabel("Price")
plt.ylabel("Count")
plt.show()

# Boxplot
plt.figure(figsize=(8,4))
plt.boxplot(df["price"], vert=False)
plt.title("House Price Boxplot")
plt.show()
```

```python
# IQR Method for Outlier Removal

Q1 = df["price"].quantile(0.25)
Q3 = df["price"].quantile(0.75)
IQR = Q3 - Q1

lower_limit = Q1 - 1.5 * IQR
upper_limit = Q3 + 1.5 * IQR

print("Lower Limit:", lower_limit)
print("Upper Limit:", upper_limit)

print("Original Dataset Shape:", df.shape)

df_clean = df[(df["price"] >= lower_limit) & (df["price"] <= upper_limit)]

print("Clean Dataset Shape:", df_clean.shape)

print("Outliers Removed:", len(df) - len(df_clean))
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
plt.hist(df_clean["price"], bins=50)
plt.title("Price Distribution After Outlier Removal")
plt.xlabel("Price")
plt.ylabel("Count")
plt.show()

plt.figure(figsize=(8,4))
plt.boxplot(df_clean["price"], vert=False)
plt.title("Boxplot After Outlier Removal")
plt.show()
```

```python
import geopandas as gpd
from shapely.geometry import Point

# Create Geometry
geometry = [Point(xy) for xy in zip(df_clean["long"], df_clean["lat"])]

# Create GeoDataFrame
gdf = gpd.GeoDataFrame(df_clean, geometry=geometry)

# Set Coordinate Reference System
gdf.set_crs(epsg=4326, inplace=True)

print(gdf.head())

print("\nCRS:", gdf.crs)
print("Shape:", gdf.shape)
```

```python
import folium

# Center Map
center_lat = gdf["lat"].mean()
center_long = gdf["long"].mean()

m = folium.Map(
    location=[center_lat, center_long],
    zoom_start=10
)

# Show first 500 houses (performance ke liye)
for _, row in gdf.head(500).iterrows():
    folium.CircleMarker(
        location=[row["lat"], row["long"]],
        radius=2,
        popup=f'Price: ${row["price"]}',
        color="blue",
        fill=True,
        fill_opacity=0.6
    ).add_to(m)

m
```

```python
m.save("../dashboard/house_price_map.html")
```

```python
df_clean.to_csv("../data/processed/kc_house_clean.csv", index=False)

print("Processed dataset saved successfully.")
```

```python
from sklearn.preprocessing import StandardScaler

# Columns to normalize
columns = [
    "bedrooms",
    "bathrooms",
    "sqft_living",
    "sqft_lot",
    "floors"
]

# Create scaler
scaler = StandardScaler()

# Make a copy of cleaned data
df_normalized = df_clean.copy()

# Normalize selected columns
df_normalized[columns] = scaler.fit_transform(df_normalized[columns])

# Display normalized values
df_normalized[columns].head()
```

```python
df_normalized.to_csv("../data/processed/kc_house_normalized.csv", index=False)

print("Normalized dataset saved successfully!")
```
